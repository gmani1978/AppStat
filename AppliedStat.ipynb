# Question 2

A LED bulb manufacturing company regularly conducts quality checks at specified periods on the products it manufactures. 
Historically, the failure rate for LED light bulbs that the company manufactures is 5%. Suppose a random sample of 10 LED light 
bulbs is selected. What is the probability that 

a) None of the LED bulbs are defective?

b) Exactly one of the LED bulbs is defective?

c) Two or fewer of the LED bulbs are defective?

d) Three or more of the LED bulbs are defective

```python
import numpy             as np
import pandas            as pd
import scipy.stats       as stats
import matplotlib.pyplot as plt
```

```python
p2   =  0.05 # failure rate for LED light bulbs that the company manufactures is 5%
n2   =  10 # sample size
k2   =  np.arange(0,11) # An array of quantiles
```

```python
binom2 = stats.binom.pmf(k2,n2,p2)
```

```python
print(binom2)
```

```python
print("None of the LED bulbs are defective?")
print(binom2[0])
```

```python
print('Exactly one of the LED bulbs is defective? is')
print(binom2[1])
```

To answer the question c) Two or fewer of the LED bulbs are defective?

we need to calculate cumulative probability of Upto Two LED bulbs being defective

```python
cum = stats.binom.cdf(k2,n2,p2)
```

```python
print(cum)
```

```python
print('Two or fewer of the LED bulbs are defective? is')
cum[2]
```

To answer the question d) Three or more of the LED bulbs are defective, we need to subtract cumulative Probability upto 2 
defective LED bulbs from 1.

```python
P = 1- cum[2]
print('Three or more of the LED bulbs are defective is')
print(P)
```

# Question 3

A car salesman sells on an average 3 cars per week. A. Probability that in a given week he will sell some cars. B. Probability that in a given week he will sell 2 or more but less than 5 cars.

Here $\lambda$ = 3

```python
rate3 = 3
n3 = np.arange(10)
p3 = stats.poisson.cdf(n3,rate3)
p3
```

```python
# P = probability that the agent sells some cars
Somecars   = 1 - p3[0]
print('Probability that the agent sells some cars is')
print(Somecars)
```

b. Agent sells 2 or more but less than 5 cars

```python
# Let P1 be the probability that the agent sells more than 4 cars
MoreThanFour = p3[4] # P(X >=4)
# Let P2 be the probability that the agent sells more than 1 car
MoreThanOne = p3[1] # P(X >=1)
TwoFive = MoreThanFour - MoreThanOne # Prob. that the agent will sell between 2 and 5 cars
print(' b) Probability that the agent sells between 2 and 5 cars is')
print(TwoFive)
```

- Visually the distributions are quiet similar
- To prove if there is a significant difference or not we will have to perform a hypothesis test
- Since we want to compare the difference in means of the samples, we can either use 2 sample t test or z-test
- For this exercise we shall do both and compare the results
- One thing to notice is that the distributions are not normal yet, since our sample size is reasonably big ( >30 samples), we will go ahead with the aforementioned tests



# For the first test, 

# Null hypothesis (Ho)
    
- There is no difference between the two means

# **Alternate hypothesis (Ha)** : 

- There is significant difference between the two means

```python
from statsmodels.stats.weightstats import ztest

operating = data1.loc[data1.OperatingState == 'Operating', 'Funds_in_million']
closed = data1.loc[data1.OperatingState =='Closed', 'Funds_in_million']

alpha = 0.05   # Let's consider a significance level of 5% - industary standard

test_statistic, pvalue = ztest(operating, closed)

if pvalue <= alpha:
    print(f'Since the p-value, {round(pvalue, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(pvalue,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
    
```

```python
from scipy.stats import ttest_ind

test_statistic, pvalue = ttest_ind(operating, closed)

if pvalue <= alpha:
    print(f'Since the p-value, {round(pvalue, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(pvalue,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

- From the tests above we found no evidence to state that the companies that have raised more money tend to suceed more or vice-versa

- There is some difference between the proportion operational companies in the two categories, but let's see if the differnce is significant
- There are two ways we can check if the difference is significant
    - Z-test for proportions
    - Chi Square Test
- We shall perform z-test in this case and compare with Chi Square to compare the hypothesis testing

<span style="font-family: Arial; font-weight:bold;font-size:1em;color:#5B2C6F">**Null hyputhesis (Ho)**: The proportion of companies that are operating is the same in both categories - winners and contestants

<span style="font-family: Arial; font-weight:bold;font-size:1em;color:#5B2C6F">**Alternative hypothesis (Ha)**: The proportion of companies that are operating is significantly different from each other, among the two categories

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#5B2C6F">**Z-Test**

```python
from statsmodels.stats.proportion import proportions_ztest

test_statistic, p_value = proportions_ztest([contestants_operating, winners_operating], [contestants, winners])

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#5B2C6F">**Chi Square-Test**

```python
from statsmodels.stats.proportion import proportions_chisquare
import numpy as np

stat, p_value, test = proportions_chisquare ([contestants_operating, winners_operating], [contestants, winners])

if p_value <= alpha:
   print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
   print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,\t we fail to reject the Null hypothesis''')
```
